```python
# generating linear looking data to test the linear model's equation
import numpy as np

x = 2*np.random.rand(100,1)
y = 4 + 3*x + np.random.randn(100,1)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.axis([0,2,0,15])
ax.plot(x,y,'o')
```

### numpy c_ practice :Translates slice objects to concatenation along the second axis.

```python
arr = np.random.rand(4,1)
arr2 = np.ones((4,1))
arr.shape, arr2.shape
```

```python
arr
```

```python
np.c_[arr2,arr].shape
```

```python
# compute theta-hat using the Normal equation
x_b = np.c_[np.ones((100,1)),x]  # adding x0 = 1 to each instance
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
theta_best
```

```python
x_new = np.array([[0],[2]])
x_new_b = np.c_[np.ones((2,1)),x_new]
y_predict = x_new_b.dot(theta_best)
y_predict
```

### Let's plot this model's predictions

```python
plt.plot(x_new,y_predict,"r-",label="Predictions")
plt.plot(x,y,"b.")
plt.axis([0,2,0,15])
plt.legend(loc="upper left")
plt.show()
```
